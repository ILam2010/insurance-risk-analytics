## Missing Value Assessment and Handling

A data quality assessment was conducted to identify missing values across all variables in the insurance dataset. Several columns contained missing observations, with the severity of missingness varying significantly between features.

Columns such as `NumberOfVehiclesInFleet`, `CrossBorder`, `WrittenOff`, `Rebuilt`, and `Converted` exhibited extremely high levels of missing data, reducing their analytical reliability. Features with excessive missingness provide limited predictive value and may introduce bias or instability into downstream analysis and modeling.

For numerical variables with moderate missingness, median imputation was selected instead of mean imputation because insurance-related financial variables often contain extreme outliers. Median values are therefore more robust and representative.

For categorical variables such as `Gender`, `MaritalStatus`, and `Bank`, missing values were replaced with the category `"Unknown"` in order to preserve records while explicitly identifying unavailable information.

Additionally, missing values observed in the derived `LossRatio` metric were primarily caused by policies with zero premium values, leading to division-by-zero errors during calculation. These cases were safely handled by assigning a loss ratio value of zero where premium values were unavailable or equal to zero.

This approach ensures improved data consistency, minimizes information loss, and prepares the dataset for reliable exploratory analysis and predictive modeling.

```python
# Numerical columns to clean
numerical_cols = [
    "CustomValueEstimate",
    "CapitalOutstanding",
    "Cylinders",
    "cubiccapacity",
    "kilowatts",
    "NumberOfDoors"
]

# Convert columns to numeric properly
for col in numerical_cols:

    # Convert to string
    df[col] = df[col].astype(str)

    # Replace commas with dots
    df[col] = df[col].str.replace(",", ".", regex=False)

    # Convert to numeric
    df[col] = pd.to_numeric(df[col], errors="coerce")

    # Fill missing values with median
    df[col] = df[col].fillna(df[col].median())
```

```python
df.isnull().sum()
```

```python
df.drop(columns=[
    "WrittenOff",
    "Rebuilt",
    "Converted",
    "NewVehicle"
], inplace=True)
```

```python
vehicle_cols = [
    "mmcode", "VehicleType", "make", "Model",
    "bodytype", "VehicleIntroDate"
]

for col in vehicle_cols:
    df[col] = df[col].fillna("Unknown")
```

```python
import numpy as np

df["LossRatio"] = np.where(
    df["TotalPremium"] > 0,
    df["TotalClaims"] / df["TotalPremium"],
    0
)
```

## Final Missing Value Treatment Summary

After initial data profiling, missing values were handled using a combination of removal, imputation, and categorical grouping strategies depending on the severity and nature of missingness.

Features with extremely high missing rates (e.g., WrittenOff, Rebuilt, Converted, NewVehicle) were removed due to their limited analytical value and potential to introduce bias.

Vehicle-related variables with low missingness were imputed using the category "Unknown" to preserve records while maintaining interpretability.

Numerical variables were cleaned and standardized before applying median-based imputation, ensuring robustness against extreme outliers.

Finally, derived metrics such as Loss Ratio were recalculated using a safe division approach to prevent invalid or infinite values caused by zero-premium policies.
